<h1 align="center"> T20 World Cup Cricket Data Pre Processing </h1>

```python
import pandas as pd
import json
```

Process Match Results

```python
with open('/content/t20_wc_match_results.json') as f:
  data = json.load(f)

df_match =pd.DataFrame(data[0]['matchSummary'])
df_match.head()
```

```python
df_match.shape
```

**Use scorecard as a match id to link with other tables**

```python
df_match.rename({'scorecard':'match_id'},axis =1,inplace =True)
df_match.head()
```

**Create a match ids dictionary that maps team names to a unique match id. This will be useful later on to link with other tables**

```python
match_id_dict ={}

for index, row in df_match.iterrows():
  key1 = row['team1'] + ' Vs ' + row['team2']
  key2 = row['team2'] + ' Vs ' + row['team1']

  match_id_dict[key1]= row["match_id"]
  match_id_dict[key2]= row["match_id"]

match_id_dict
```

```python
df_match.to_csv('/content/CsvFile/dim_match_summary.csv', index = False)
```

Process Batting Summary

```python
with open('/content/t20_wc_batting_summary.json') as f:
  data = json.load(f)

  all_records=[]

  for rec in data:
    all_records.extend(rec['battingSummary'])
df_batting = pd.DataFrame(all_records)
df_batting.head()
```

```python
df_batting["out/not_out"]= df_batting.dismissal.apply(lambda x: "out" if len(x)>0 else "not_out")
df_batting.head()
```

```python
df_batting.drop(columns =["dismissal"],inplace = True)
df_batting.head(10)
```

**Cleanup weird characters**

```python
df_batting['batsmanName']= df_batting['batsmanName'].apply(lambda x: x.replace('†',''))
df_batting['batsmanName']= df_batting['batsmanName'].apply(lambda x: x.replace('(c)',''))
df_batting.head(10)
```

```python
df_batting["match_id"]= df_batting["match"].map(match_id_dict)
df_batting.head()
```

```python
df_batting.to_csv('/content/CsvFile/fact_bating_summary.csv', index = False)
```

Bowling Summary

```python
with open('/content/t20_wc_bowling_summary.json') as f:
    data = json.load(f)
    all_records = []
    for rec in data:
        all_records.extend(rec['bowlingSummary'])
all_records[:2]
```

```python
df_bowling = pd.DataFrame(all_records)
print(df_bowling.shape)
df_bowling.head()
```

```python
df_bowling['match_id'] = df_bowling['match'].map(match_id_dict)
df_bowling.head()
```

```python
df_bowling.to_csv('/content/CsvFile/fact_bowling_summary.csv', index = False)
```

Process Players Information

```python
with open('/content/t20_wc_player_info.json') as f:
    data = json.load(f)
```

```python
df_players = pd.DataFrame(data)

print(df_players.shape)
df_players.head(10)
```

Cleanup Weird Characters

```python
df_players['name'] = df_players['name'].apply(lambda x: x.replace('â€', ''))
df_players['name'] = df_players['name'].apply(lambda x: x.replace('†', ''))
df_players['name'] = df_players['name'].apply(lambda x: x.replace('\xa0', ''))
df_players.head(10)
```
